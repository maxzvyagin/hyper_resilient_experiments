```python
from hyper_resilient_experiments.simple_mnist import pt_mnist, tf_mnist
```

```python
def objective(config):
    average_res, model = pt_mnist.mnist_pt_objective(config)
    search_results = {}
    search_results['average_res'] = average_res
    tune.report(**search_results)
    return search_results
```

```python
class Namespace:
    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)
```

```python
import inspect
inspect.getsource(create_hyperspace)
```

```python
args = Namespace(json='/Users/mzvyagin/Documents/resiliency/dependency/standard.json', trials=10, out='test.csv')
try:
    ray.init()
except:
    pass
space, bounds = spaceray.get_trials(args)
```

```python
optimizer = Optimizer(space[0], random_state=0)
search_algo = SkOptSearch(optimizer, list(bounds.keys()), metric='average_res', mode='min')
```

```python
analysis = tune.run(objective, search_alg=search_algo, num_samples=int(args.trials))
```
